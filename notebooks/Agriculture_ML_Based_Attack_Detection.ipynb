Dataset spliting For model Training and Testing Purpose

```python
from sklearn.model_selection import train_test_split
Features_columns=df.iloc[:,:-1]
Target_column=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(Features_columns,Target_column,test_size=0.2,random_state=42,stratify=Target_column)
```

Applying Model BuildIng To the Preproceesd Dataset

Starting with model 1 :

**LogisticRegression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regrestion_model=LogisticRegression(max_iter=1000)
```

```python
logistic_regrestion_model.fit(x_train,y_train)
```

```python
x_train.dtypes
```

```python
# Preprocess string columns
from sklearn.preprocessing import LabelEncoder
le_X = LabelEncoder()

# Check values for object columns
for col in ['WaterPumpToTank', 'WaterPumpFromTank', 'Light']:
    if col in x_train.columns:
        x_train[col] = le_X.fit_transform(x_train[col])
        x_test[col] = le_X.transform(x_test[col])

# Drop Timestamp
if 'Timestamp' in x_train.columns:
    x_train = x_train.drop('Timestamp', axis=1)
if 'Timestamp' in x_test.columns:
    x_test = x_test.drop('Timestamp', axis=1)

# Ensure y_train and y_test are numeric
le_y = LabelEncoder()
y_train = le_y.fit_transform(y_train)
y_test = le_y.transform(y_test)

x_train.dtypes
```

```python
predicted_values_logistic_regrestion=logistic_regrestion_model.predict(x_test)
```

```python
predicted_values_logistic_regrestion
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, roc_auc_score,precision_score,recall_score, matthews_corrcoef

print(f"The Model Scores For the Logistic Regression")
print()
print(f"Accuracy: {accuracy_score(y_test, predicted_values_logistic_regrestion):.4f}")
print(f"F1 Score: {f1_score(y_test, predicted_values_logistic_regrestion):.4f}")
print(f"precision_scoe: {f1_score(y_test, predicted_values_logistic_regrestion):.4f}")
print(f"recall_score: {f1_score(y_test, predicted_values_logistic_regrestion):.4f}")
print(f"ROC-AUC: {roc_auc_score(y_test, predicted_values_logistic_regrestion):.4f}")
print(f"Matthews Corr. Coef.: {matthews_corrcoef(y_test, predicted_values_logistic_regrestion):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, predicted_values_logistic_regrestion))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, predicted_values_logistic_regrestion))
```

```python
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    StackingClassifier
)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report,
    matthews_corrcoef
)

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# ==========================================
# Base Models
# ==========================================

rf = RandomForestClassifier(
    random_state=42
)

et = ExtraTreesClassifier(
    random_state=42
)

xgb = XGBClassifier(
    random_state=42,
    eval_metric='logloss'
)

lgbm = LGBMClassifier(
    random_state=42
)

# ==========================================
# Meta Model
# ==========================================

meta_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

# ==========================================
# Stacking Classifier
# ==========================================

stacking_model = StackingClassifier(
    estimators=[
        ('rf', rf),
        ('et', et),
        ('xgb', xgb),
        ('lgbm', lgbm)
    ],
    final_estimator=meta_model,
    cv=5,
    n_jobs=-1 # Use all available CPU cores
)

# ==========================================
# Train Model
# ==========================================

stacking_model.fit(x_train, y_train)

# ==========================================
# Predictions
# ==========================================

y_pred = stacking_model.predict(x_test)

y_prob = stacking_model.predict_proba(x_test)[:, 1]

# ==========================================
# Evaluation Metrics
# ==========================================

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_prob)
mcc = matthews_corrcoef(y_test, y_pred)

print("Accuracy:", round(accuracy, 4))
print("Precision:", round(precision, 4))
print("Recall:", round(recall, 4))
print("F1 Score:", round(f1, 4))
print("ROC-AUC:", round(roc_auc, 4))
print("Matthews Corr. Coef.:", round(mcc, 4))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
